Script que promove a criação de um arquivo '.sql' que contém informações para a execução do dump dos dados no DataBase. Etapas:
* 1) CSV para DataFrame
* 2) o DataFrame criado é usado para a criação do arquivos '.sql', com todos os dados que devem ser instanciados

```python
import pandas as pd
```

```python
file = open('aeronautical_occurrences_database.sql', 'w')
```

```python
anv = pd.read_csv('../data/anv.csv', sep='~')
ftc = pd.read_csv('../data/ftc.csv', sep='~')
oco = pd.read_csv('../data/oco.csv', sep='~')
rec = pd.read_csv('../data/rec.csv', sep='~')
```

##### Para o CSV anv

```python
anv = anv.drop_duplicates(subset="codigo_ocorrencia", keep=False).reset_index()
anv = anv.drop(['index'], axis=1)
```

```python
# Para 'aeronave_dia_extracao' checar o link https://www.sqlitetutorial.net/sqlite-date/
file.write("CREATE TABLE [AeronavesEnvolvidas] (\n" +
           "codigo_ocorrencia INT,\n" +
           "aeronave_matricula VARCHAR(20),\n" +
           "aeronave_operador_categoria VARCHAR(60),\n" +
           "aeronave_tipo_veiculo VARCHAR(30),\n" +
           "aeronave_fabricante VARCHAR(100),\n" +
           "aeronave_modelo VARCHAR(20),\n" +
           "aeronave_tipo_icao VARCHAR(20),\n" +
           "aeronave_motor_tipo VARCHAR(30),\n" +
           "aeronave_motor_quantidade VARCHAR(20), \n" +
           "aeronave_pmd INT,\n" +
           "aeronave_pmd_categoria VARCHAR(50),\n" +
           "aeronave_assentos INT,\n" +
           "aeronave_ano_fabricacao INT,\n" +
           "aeronave_pais_fabricante VARCHAR(30),\n" +
           "aeronave_pais_registro VARCHAR(30),\n" +
           "aeronave_registro_categoria VARCHAR(20),\n" +
           "aeronave_registro_segmento VARCHAR(20),\n" +
           "aeronave_voo_origem VARCHAR(20),\n" +
           "aeronave_voo_destino VARCHAR(20),\n" +
           "aeronave_fase_operacao VARCHAR(50),\n" +
           "aeronave_fase_operacao_icao VARCHAR(30),\n" +
           "aeronave_tipo_operacao VARHCAR(40),\n" +
           "aeronave_nivel_dano VARCHAR(30),\n" +
           "total_fatalidades INT,\n" +
           "aeronave_dia_extracao TEXT,\n" +
           "PRIMARY KEY (codigo_ocorrencia)\n"+
           ");\n\n")
```

```python
for i in anv.index:
    file.write("INSERT INTO AeronavesEnvolvidas " +
               
               "(codigo_ocorrencia, aeronave_matricula, " +
               "aeronave_operador_categoria, aeronave_tipo_veiculo, " +
               "aeronave_fabricante, aeronave_modelo, aeronave_tipo_icao, " +
               "aeronave_motor_tipo, aeronave_motor_quantidade, aeronave_pmd, " +
               "aeronave_pmd_categoria, aeronave_assentos, " +
               "aeronave_ano_fabricacao, aeronave_pais_fabricante, " +
               "aeronave_pais_registro, aeronave_registro_categoria, " +
               "aeronave_registro_segmento, aeronave_voo_origem, " +
               "aeronave_voo_destino, aeronave_fase_operacao, " +
               "aeronave_fase_operacao_icao, aeronave_tipo_operacao, " +
               "aeronave_nivel_dano, total_fatalidades, aeronave_dia_extracao) " +
               
               "VALUES " +
               
               "("+str(anv.iloc[i]['codigo_ocorrencia'])+", '"+str(anv.iloc[i]['aeronave_matricula'])+"', '"+
               str(anv.iloc[i]['aeronave_operador_categoria'])+"', '"+str(anv.iloc[i]['aeronave_tipo_veiculo'])+"', '"+
               str(anv.iloc[i]['aeronave_fabricante'])+"', '"+str(anv.iloc[i]['aeronave_modelo'])+"', '"+str(anv.iloc[i]['aeronave_tipo_icao'])+"', '"+
               str(anv.iloc[i]['aeronave_motor_tipo'])+"', '"+str(anv.iloc[i]['aeronave_motor_quantidade'])+"', "+str(anv.iloc[i]['aeronave_pmd'])+", '"+
               str(anv.iloc[i]['aeronave_pmd_categoria'])+"', "+str(anv.iloc[i]['aeronave_assentos'])+", "+
               str(anv.iloc[i]['aeronave_ano_fabricacao'])+", '"+str(anv.iloc[i]['aeronave_pais_fabricante'])+"', '"+
               str(anv.iloc[i]['aeronave_pais_registro'])+"', '"+str(anv.iloc[i]['aeronave_registro_categoria'])+"', '"+
               str(anv.iloc[i]['aeronave_registro_segmento'])+"', '"+str(anv.iloc[i]['aeronave_voo_origem'])+"', '"+
               str(anv.iloc[i]['aeronave_voo_destino'])+"', '"+str(anv.iloc[i]['aeronave_fase_operacao'])+"', '"+
               str(anv.iloc[i]['aeronave_fase_operacao_icao'])+"', '"+str(anv.iloc[i]['aeronave_tipo_operacao'])+"', '"+
               str(anv.iloc[i]['aeronave_nivel_dano'])+"', "+str(anv.iloc[i]['total_fatalidades'])+", "+str(anv.iloc[i]['aeronave_dia_extracao'])+");\n")
```
